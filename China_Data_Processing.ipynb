```python
import pandas as pd
```

```python
csv_file = pd.DataFrame(pd.read_csv("CN_Employment_Data.csv", sep = ",", header = 0))
df1_transposed = csv_file.T
df1_transposed.columns = ['Employed population', 'Employed Population in City', 'Unemployment rate']
```

```python
csv_file.to_json("CN_Employment.json", orient = "records", date_format = "epoch", double_precision = 10, force_ascii = True, date_unit = "ms", default_handler = None)
```

```python
CN_employment_data = df1_transposed.drop("Unnamed: 0")
```

```python
csv_file2 = pd.DataFrame(pd.read_csv("CN_Province_Unemployment_Rate.csv", sep = ",", header = 0))
df2_transposed = csv_file2.T
city_name = csv_file2.iloc[:, 0].tolist()
df2_transposed.columns = city_name
CN_unemployment_data = df2_transposed.drop("Unnamed: 0")
```
